## 2. Perform image processing
In this step, we perform image processing, which is a process of matching multiband imaging data in both spatial resolution (i.e., PSF size) and sampling (i.e., pixel size). Basically, this process includes PSF matching and spatial resampling and reprojection to the multiband images. This image processing can be done using the `piXedfit.piXedfit_images.images_processing` class.

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# call images_processing class
from piXedfit.piXedfit_images import images_processing
```

### 2.1. Preparing inputs

```python
# list the filters
filters = ['galex_fuv', 'galex_nuv', 'sdss_u', 'sdss_g', 'sdss_r', 'sdss_i', 
           'sdss_z', '2mass_j', '2mass_h', '2mass_k', 'wise_w1', 'wise_w2']

# input science images
sci_img = {}
sci_img['galex_fuv'] = 'MISWZS01_29057_0266-fd-intbgsub.fits'
sci_img['galex_nuv'] = 'MISWZS01_29057_0266-nd-intbgsub.fits'
sci_img['sdss_u'] = 'frame-u-004263-6-0215.fits'
sci_img['sdss_g'] = 'frame-g-004263-6-0215.fits'
sci_img['sdss_r'] = 'frame-r-004263-6-0215.fits'
sci_img['sdss_i'] = 'frame-i-004263-6-0215.fits'
sci_img['sdss_z'] = 'frame-z-004263-6-0215.fits'
sci_img['2mass_j'] = 'skybgsub_aJ_asky_000816s0630044.fits'
sci_img['2mass_h'] = 'skybgsub_aH_asky_000816s0630044.fits'
sci_img['2mass_k'] = 'skybgsub_aK_asky_000816s0630044.fits'
sci_img['wise_w1'] = 'skybgsub_0151p015_ac51-w1-int-3_ra14.925546_dec0.916659_asec1000.000.fits'
sci_img['wise_w2'] = 'skybgsub_0151p015_ac51-w2-int-3_ra14.925546_dec0.916659_asec1000.000.fits'

# input Variance images
var_img = {}
var_img['galex_fuv'] = 'var_MISWZS01_29057_0266-fd-intbgsub.fits'
var_img['galex_nuv'] = 'var_MISWZS01_29057_0266-nd-intbgsub.fits'
var_img['sdss_u'] = 'var_frame-u-004263-6-0215.fits'
var_img['sdss_g'] = 'var_frame-g-004263-6-0215.fits'
var_img['sdss_r'] = 'var_frame-r-004263-6-0215.fits'
var_img['sdss_i'] = 'var_frame-i-004263-6-0215.fits'
var_img['sdss_z'] = 'var_frame-z-004263-6-0215.fits'
var_img['2mass_j'] = 'var_skybgsub_aJ_asky_000816s0630044.fits'
var_img['2mass_h'] = 'var_skybgsub_aH_asky_000816s0630044.fits'
var_img['2mass_k'] = 'var_skybgsub_aK_asky_000816s0630044.fits'
var_img['wise_w1'] = 'var_0151p015_ac51-w1-unc-3_ra14.925546_dec0.916659_asec1000.000.fits'
var_img['wise_w2'] = 'var_0151p015_ac51-w2-unc-3_ra14.925546_dec0.916659_asec1000.000.fits'

gal_ra = 14.925546
gal_dec = 0.916659
gal_z = 0.017636

# size of the final stamp images
stamp_size = [101,101]
```

### 2.2. Call images_processing class

```python
img_process = images_processing(filters=filters,sci_img=sci_img,var_img=var_img,gal_ra=gal_ra,gal_dec=gal_dec,
                                gal_z=gal_z,stamp_size=stamp_size)
```

### Run image processing

```python
output_stamps = img_process.reduced_stamps()
```

### Check results of images processing: stamps of reduced images

```python
%matplotlib inline

fig1 = plt.figure(figsize=(20,7))

nbands = len(filters)
for bb in range(0,nbands):
    f1 = fig1.add_subplot(2, 6, bb+1)
    plt.tick_params(left=False, right=False, labelleft=False, labelbottom=False, bottom=False)
    str_temp = "name_img_%s" % filters[bb]
    hdu = fits.open(output_stamps[str_temp])
    plt.imshow(np.log10(hdu[0].data), origin='lower')
    f1.text(0.5, 0.93, filters[bb], horizontalalignment='center', 
            verticalalignment='center',transform = f1.transAxes, 
            fontsize=22, color='black')
    hdu.close()

plt.subplots_adjust(left=0.05, right=0.95, bottom=0.05, top=0.95, hspace=0.05, wspace=0.05)
```

### 2.3. Define galaxy's region of interest
Basically, user has a freedom to define the galaxy's region of interest (ROI) that will be analyzed further (i.e., SED fitting). The input ROI should be in a 2D array with the same size as that of the stamp images produced after the image processing. In the 2D array, pixels belong to the galaxy should be given value 1 and those not belong to the galaxy should be given value 0.

While it can be freely defined, It's recmmended to use a built-in function in piXedfit for defining the galaxy's region. There are two options provided for this: (1) using segmentation maps, and (2) using elliptical aperture. Here we will demnstrate these two options, but at the end we will use the ROI defined based on segmentation maps

### 2.3.1. Based on segmentation maps

This can be done using `segmentation_sep` function.

```python
segm_maps = img_process.segmentation_sep(output_stamps, thresh=2.8, minarea=100, deblend_nthresh=60, 
                                         deblend_cont=0.001)
```

```python
fig1 = plt.figure(figsize=(20,7))

for bb in range(0,nbands):
    f1 = fig1.add_subplot(2, 6, bb+1)
    plt.tick_params(left = False, right = False , labelleft = False ,
                    labelbottom = False, bottom = False)
    plt.imshow(segm_maps[bb], origin='lower')
    f1.text(0.5, 0.93, "id %d: %s" % (bb,filters[bb]), horizontalalignment='center', 
            verticalalignment='center',transform = f1.transAxes, 
            fontsize=20, color='white')

plt.subplots_adjust(left=0.05, right=0.95, bottom=0.05, top=0.95, hspace=0.05, wspace=0.05)
```

Select segmentation maps that are going too be used and merge them to form a galaxy's ROI. Let's called it `gal_region1`. This can be done using the `galaxy_region` method.

```python
# select segmentation maps
select_ids = [5, 6]
select_segm_maps = []
for ii in select_ids:
    select_segm_maps.append(segm_maps[ii])

gal_region1 = img_process.galaxy_region(segm_maps=select_segm_maps)

# plot
fig1 = plt.figure(figsize=(5,5))
f1 = plt.subplot()

str_temp = "name_img_%s" % filters[3]
hdu = fits.open(output_stamps[str_temp])
data_image = hdu[0].data
hdu.close()
plt.imshow(np.log10(data_image), origin='lower')
    
plt.imshow(gal_region1, origin='lower', cmap='Greys', alpha=0.2)
```

### 2.3.2. Based on simple elliptical aperture

This can also be performed using `galaxy_region`. Let's use a circular aperture (ellipticity=0) and call the region defined with this method as `gal_region2`.

```python
# ellipse center
y_cent = (stamp_size[0]-1)/2
x_cent = (stamp_size[1]-1)/2 

# ellipticity and position angle. Let's assume circular and choose arbitrary position angle.
ell = 0.0  
pa = 45.0

# radius
radius_sma = 22.0 

gal_region2 = img_process.galaxy_region(use_ellipse=True, x_cent=x_cent, y_cent=y_cent, 
                                        ell=ell, pa=pa, radius_sma=radius_sma) 

# plot
fig1 = plt.figure(figsize=(5,5))
f1 = plt.subplot()
str_temp = "name_img_%s" % filters[3]
hdu = fits.open(output_stamps[str_temp])
data_image = hdu[0].data
hdu.close()
plt.imshow(np.log10(data_image), origin='lower')
    
plt.imshow(gal_region2, origin='lower', cmap='Greys', alpha=0.2)
```

### 2.4. Calculate E(B-V) attenuation by the foreground Galactic dust 
This value can be estimated using the `EBV_foreground_dust` function. This function retrieve the information from IRSA database through astroquery module. 

```python
from piXedfit.piXedfit_images import EBV_foreground_dust

Gal_EBV = EBV_foreground_dust(gal_ra, gal_dec)
```

### 2.5. Calculate fluxes of pixels and get photometric data cube
This can be done using the `flux map` method. We need to input galaxy's ROI to this function. For this tutorial, let's use the one we obtained based on the segmentation maps.

```python
name_out_fits = "fluxmap_12073-12703.fits"
flux_maps = img_process.flux_map(output_stamps, gal_region1, Gal_EBV=Gal_EBV,
                                 name_out_fits=name_out_fits)
```

### 2.6. Check result of image processing: photometric data cube

```python
hdu = fits.open("fluxmap_12073-12703.fits")
hdu.info()
header = hdu[0].header
print (header)

# get unit of flux in erg/s/cm2/A
unit_flux = float(header['unit'])

# get maps
gal_region = hdu['GALAXY_REGION'].data
flux_map = hdu['FLUX'].data*unit_flux
flux_err_map = hdu['FLUX_ERR'].data*unit_flux
hdu.close()
```

Get central wavelengths of the filters to be used for plotting SEDs of pixels. This can be obtained using `piXedfit.utils.filtering.cwave_filters` function.

```python
from piXedfit.utils.filtering import cwave_filters

photo_wave = cwave_filters(filters)
```

### 2.6.2. Check S/N ratio profiles

```python
# get pixels belong to the galaxy's region
rows, cols = np.where(gal_region==1)

# size of the image
dim_y, dim_x = gal_region.shape[0], gal_region.shape[1]

# estimate galactic center position
x_cent, y_cent = (dim_x-1)/2, (dim_y-1)/2

e = 0.0
pa = 45.0

# calculate radii of the pixels within the galaxy 
from piXedfit.piXedfit_images import ellipse_sma

pix_sma = ellipse_sma(e,pa,cols-x_cent,rows-y_cent)

# calculate the S/N ratios of pixels
pix_SNR = np.zeros((nbands,len(rows)))
for bb in range(0,nbands):
    pix_SNR[bb] = flux_map[bb][rows,cols]/flux_err_map[bb][rows,cols]
```

```python
fig1 = plt.figure(figsize=(10,6))
f1 = plt.subplot()
plt.ylim(-1.5,2.6)
plt.setp(f1.get_yticklabels(), fontsize=13, visible=True)
plt.setp(f1.get_xticklabels(), fontsize=13, visible=True)
plt.xlabel(r"Radius [pixel]", fontsize=18)
plt.ylabel(r"log(S/N Ratio)", fontsize=18)

for bb in range(0,nbands):
    plt.scatter(pix_sma, np.log10(pix_SNR[bb]), s=5, alpha=0.5, 
                label='%s' % filters[bb])
    
plt.legend(fontsize=12, ncol=2, loc=1)
```
